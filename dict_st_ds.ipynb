# dict
1.dict is collection of item enclosed by {}-item is key value of pair

```python
info={"name":"shiva","salary":10000.25,"adress":"hyderbad","age":25}
print(info)
print(type(info))
```

```python
# read data and updatefrom dict
info={"name":"shiva","salary":10000.25,"adress":"hyderbad","age":25}
print(info)
print(info["adress"])
```

```python
info={"name":"shiva","salary":10000.25,"adress":"hyderbad","age":25}
print(info)
info["adress"]="banglore"
print(info)
```

```python
# methods or function in dict
print(dir(dict))
```

```python
#update()
info={"name":"shiva","salary":10000.25,"adress":"hyderbad","age":25}
print(info)
info.update({"salary":10000.50})
print(info)
```

```python
#pop()
info={"name":"shiva","salary":10000.25,"adress":"hyderbad","age":25}
print(info)
info.pop("salary")
print(info)
```

# sets
1.set is a collection of unorder unique elements enclosed by {}

```python
# create set
a={1,2,3,5,6,7,8,5,4}
print(a)
print(type(a))
```

```python
# methods or functions
print(dir(set))
```

```python
# add()
a={20,30,10,40}
print(a)
a.add(100)
print(a)
```

```python
#difference()
a={1,2,3,4,5,6}
b={4,5,6,7,8,9}
c=a.difference(b)
print(c)
```

```python
a={1,2,3,4,5,6}
b={4,5,6,7,8,9}
c=a.union(b)
print(c)
```

```python
#intersection
a={1,2,3,4,5,6}
b={4,5,6,7,8,9}
c=a.intersection(b)
print(c)
```

```python
#discord
a={1,2,3,4,5,6}
print(a)
a.discard(6)
print(a)
```
