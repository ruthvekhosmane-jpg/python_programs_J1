# list in data structures
1. list is collection of elements enclosed by []
2. list a mutablelis - editable
3. it's allowed duplicate elements
4. each element seperated by ,

```python
# create list 
names=["shiva","vikram","vikas","divya"]
ages=[25,26,27,25,8,29,7]
print(names)
print(ages)
print(type(names))
print(type(ages))
```

```python
#concutenation
a=[10,20,30,40,50]
b=[40,50,60,70,80,]
result=a+b
print(result)
```

```python
# repeating list
ages=[10,20,30,35,45]
print(ages*45)
```

```python
# accesing elements from list - index
ages=[15,16,17,18,19,25,26]
print(ages)
print(ages[4])
```

```python
#methods or function in list
print(dir(list))
```

```python
#description
print(help(list))
```

```python
# add elements to list append(),extend(),insert()
a=[10,20,30,40,50]
print(a)
a.extend([100,200,300])
print(a)
```

```python
a=[10,20,30,40,50]
print(a)
a.insert(1,100)
print(a)
```

```python
#remove elements from list - pop(),remove(),clear()
a=[10,20,30,40,50]
print(a)
a.pop(4)
print(a)
```

```python
a=[10,20,30,40,50]
print(a)
a.remove(50)
print(a)
```

```python
# sorting techiniques - sort(reverse-true or reverse=false)
a=[6,4,5,3,1,2]
print(a)
a.sort(reverse=True) # true means descending , false mean acending
print(a)
```
