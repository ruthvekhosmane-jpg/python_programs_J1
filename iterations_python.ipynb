````python
#write a program to find the even numbers in between 100 to 150 using for loop?
for n in range (100,151):
    if (n%2==0):
        print(n,end=",")
    
````

````python
#write a program to find the odd numbers in between 100 to 150 using for loop?
for n in range (100,151):
    if (n%2==1):
        print(n,end=",")
````

````python
#write a program to find the vowels from string using for loop?
name=str(input("please enter your name here:"))
for char in name:
    if (char in "aeiouAEIOU"):
        print(char)
        
````

````python
#write a program to find the non-vowels from string using for lopp?
name=str(input("please enter your name here:"))
for char in name:
         if (char not in "aeiouAEIOU"):
             print(char)
````

````python
#write a program to print sum of all even numbers and odd numbers using for loop and while loop?
se=0
so=0
for n in range(1,101):
    if (n%2==0):
        se=se+n
    elif (n%2==1):
        so=so+1
print("sum of even numbers in between 1 to 100 is",se)
print("sum of odd numbers in between 1 to 100 is",so)
````
