````python
#write a program to find the even numbers in between 100 to 150 using for loop?
for n in range (100,151):
    if (n%2==0):
        print(n,end=",")
    
````

````python
#write a program to find the odd numbers in between 100 to 150 using for loop?
for n in range (100,151):
    if (n%2==1):
        print(n,end=",")
````
