````python
# write a program to print sum of two numbers using function
def add_number():
    a=int(input("enter first number"))
    b=int(input("enter second number"))
    result=a+b
    print(result)
    return result
add_number()    
````

````python
#write a program to print addition of two string using function
def add_string():
    a=str(input("enter first string"))
    b=str(input("enter second string"))
    result=a+b
    print(result)
    return result
add_string()    
````

````python
# write a program to print profile data using functions with arguments
def profile(name,salary,address):
    print("my name is",name)
    print("my salary is",salary)
    print("my address is",address)
    return
profile(name="shiva",salary=10000,address="hyderbad")
profile(name="chandra",salary=25000,address="banglore")
````
